```python
import pandas as pd
```

```python
calendar_data = pd.read_csv(r'C:\Users\prith\Downloads\airbnb-data-pipeline\data\raw\calendar.csv.gz')
```

```python
print(calendar_data.shape)
```

```python
print(calendar_data.head().to_string())
```

```python
calendar_data['available'] = calendar_data['available'].map({'t':1, 'f':2})
#here i have changed t/f as 1/0 so that i can make a column availability_Rate and i just have to find mean and multiply it by 100 to get availability_rate %
```

```python
calendar_agg = calendar_data.groupby('listing_id').agg(
    availability_rate = ('available', 'mean'),
    total_days = ('available', 'count')
).reset_index()

calendar_agg['availability_rate'] = calendar_agg['availability_rate']*100
```

```python
print(calendar_agg.shape)
print(calendar_agg.head())
```

```python
calendar_agg.to_csv(r'C:\Users\prith\Downloads\airbnb-data-pipeline\data\processed\calendar_clean.csv', index=False)
```
